```python
#EMOTION DETECTION

#OMONDI MILTON C025-01-1234/2019
#NJUGUNA IAN C025-01-1197/2019

#Change directories where necessary......

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from keras.layers import Flatten, Dense
from keras.models import Model
from tensorflow.keras.utils import img_to_array, load_img
from keras.preprocessing.image import ImageDataGenerator
from keras.applications.mobilenet import MobileNet, preprocess_input 
from keras.losses import categorical_crossentropy
```

```python
# Working with pre-trained model (MobileNet) 

base_model = MobileNet( input_shape=(160,160,3), include_top= False )

for layer in base_model.layers:
  layer.trainable = False


x = Flatten()(base_model.output)
x = Dense(units=3 , activation='softmax' )(x)

# creating our model.
model = Model(base_model.input, x)
```

```python
model.compile(optimizer='adam', loss= categorical_crossentropy , metrics=['accuracy'] )
```

```python
# Import the os module
import os

# Get the current working directory
os.getcwd()
```

```python
#preparing data using data generator
train_datagen = ImageDataGenerator(
     zoom_range = 0.2, 
     shear_range = 0.2, 
     horizontal_flip=True, 
     rescale = 1./255
)

train_data = train_datagen.flow_from_directory(directory= "F:\\Notes\\IT\\IT 4\\4.2\\Data mining\\mydata\\emotion\\train", 
                                               target_size=(160,160), 
                                               batch_size=32,
                                  )


train_data.class_indices
```

```python
val_datagen = ImageDataGenerator(rescale = 1./255 )

val_data = val_datagen.flow_from_directory(directory= "F:\\Notes\\IT\\IT 4\\4.2\\Data mining\\mydata\\emotion\\train", 
                                           target_size=(160,160), 
                                           batch_size=32,
                                  )
```

```python
# to visualize the images in the traing data denerator 

t_img , label = train_data.next()

# function call to plot the images 
def plotImages(img_arr, label):
  """
  input  :- images array 
  output :- plots the images 
  """
  count = 0
  for im, l in zip(img_arr,label) :
    plt.imshow(im)
    plt.title(im.shape)
    plt.axis = False
    plt.show()
    
    count += 1
    if count == 10:
      break

# function call 
plotImages(t_img, label)
```

```python
## having early stopping and model check point 

from keras.callbacks import ModelCheckpoint, EarlyStopping

# early stopping
es = EarlyStopping(monitor='val_accuracy', min_delta= 0.01 , patience= 5, verbose= 1, mode='auto')

# model check point
mc = ModelCheckpoint(filepath="best_model.h5", monitor= 'val_accuracy', verbose= 1, save_best_only= True, mode = 'auto')

# puting call back in a list 
call_back = [es, mc]
```

```python
hist = model.fit_generator(train_data, 
                           steps_per_epoch= 10, 
                           epochs= 30, 
                           validation_data= val_data, 
                           validation_steps= 8, 
                           callbacks=[es,mc])
```

```python
# Loading the best fit model 
from keras.models import load_model
model = load_model("F:\\Notes\\IT\\IT 4\\4.2\\Data mining\\mydata\\emotion\\best_model.h5")
```

```python
h =  hist.history
h.keys()
```

```python
plt.plot(h['accuracy'])
plt.plot(h['val_accuracy'] , c = "red")
plt.title("acc vs v-acc")
plt.show()
```
